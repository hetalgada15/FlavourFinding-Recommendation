```python
import pandas as pd

# Load the datasets
#recipe_sample = pd.read_csv('recipe_sample.csv')
restaurants = pd.read_csv('restaurants.csv')

# Display the first few rows of each dataset to understand the structure

print("\nRestaurants:")
restaurants.head()


```

```python
# Profile the dataset
print("\nRestaurants Info:")
restaurants.info()
```

```python
# 1. Handle Missing Values
# Substitute missing values with "unknown"
restaurants.fillna("unknown", inplace=True)

# 2. Drop Duplicates
restaurants.drop_duplicates(inplace=True)

# 3. Drop Unnecessary Columns
columns_to_drop = ['Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'asapDeliveryAvailable', 'asapDeliveryTimeMinutes',	'asapPickupMinutes', 'pickupAvailable', 'asapPickupAvailable']
restaurants.drop(columns=columns_to_drop, inplace=True)

# 4. Transform `priceRange` Column
# Convert dollar signs to words (e.g., $ to Low, $$ to Mid, etc.)
price_mapping = {
    "$": "Low",
    "$$": "Mid",
    "$$$": "High",
    "$$$$": "Pricey"
}
restaurants['priceRange'] = restaurants['priceRange'].map(price_mapping).fillna("unknown")

# 5. Remove $ from the `description` column
restaurants['description'] = restaurants['description'].str.replace('$', '')

# Correct data types
restaurants['latitude'] = pd.to_numeric(restaurants['latitude'], errors='coerce')
restaurants['longitude'] = pd.to_numeric(restaurants['longitude'], errors='coerce')

# Standardize formats
restaurants['market'] = restaurants['market'].str.lower()
restaurants['name'] = restaurants['name'].str.lower()
restaurants['description'] = restaurants['description'].str.lower()
restaurants['displayAddress'] = restaurants['displayAddress'].str.lower()
restaurants['street'] = restaurants['street'].str.lower()
restaurants['city'] = restaurants['city'].str.lower()
restaurants['state'] = restaurants['state'].str.lower()

```

```python
restaurants.head()
```

```python
# remove • from description column

restaurants['description'] = restaurants['description'].str.replace('•', '')
```

```python
restaurants.head()
```

```python
# Save the cleaned dataset to the current directory
restaurants.to_csv('cleaned_restaurants.csv', index=False)
print("Cleaning completed. Cleaned file saved as 'cleaned_restaurants.csv'.")
```

```python
# Load the datasets
#recipe_sample = pd.read_csv('recipe_sample.csv')
food_recipe = pd.read_csv('Food_Recipe.csv')

# Display the first few rows of each dataset to understand the structure

print("\nFood Recipes:")
food_recipe.head(5370)
```

```python
food_recipe.info()
food_recipe.nunique()
```

```python


# 1. Handle Missing Values
# Substitute missing values with placeholders for string columns
food_recipe['description'].fillna("N/A", inplace=True)
food_recipe['cuisine'].fillna("N/A", inplace=True)
food_recipe['course'].fillna("N/A", inplace=True)
food_recipe['diet'].fillna("N/A", inplace=True)
food_recipe['ingredients_name'].fillna("N/A", inplace=True)
food_recipe['ingredients_quantity'].fillna("N/A", inplace=True)

# Substitute missing values with -1 for numerical columns to indicate missing data
food_recipe['prep_time (in mins)'].fillna(-1, inplace=True)
food_recipe['cook_time (in mins)'].fillna(-1, inplace=True)

# 2. Convert `prep_time (in mins)` and `cook_time (in mins)` to integer
food_recipe['prep_time (in mins)'] = food_recipe['prep_time (in mins)'].astype(int)
food_recipe['cook_time (in mins)'] = food_recipe['cook_time (in mins)'].astype(int)

# 3. Drop Duplicates
food_recipe.drop_duplicates(subset=['name'], keep='first', inplace=True)
```

```python
# Display the cleaned dataset info
print("\nFood Recipe Dataset After Cleaning:")
food_recipe.head(5370)
```

```python
# Save the cleaned dataset to the current directory
food_recipe.to_csv('cleaned_food_recipe.csv', index=False)
print("Cleaning completed. Cleaned file saved as 'cleaned_food_recipe.csv'.")
```
